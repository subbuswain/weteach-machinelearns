```python
with open('./../Datasets/customers.csv', 'r') as csv_file:
    customers = csv.DictReader(csv_file)
    fields = ['Customer Id', 'First Name', 'Last Name', 'Company', 'Country', 'Email']
    with open('./new_file.csv', 'w') as newfile:
        writer = csv.DictWriter(newfile, fieldnames=fields, delimiter=' ')
        writer.writeheader()
        for customer in customers:
            write_data = {key: value for key, value in customer.items() if key in fields} # comprehension
            writer.writerow(write_data)
```
